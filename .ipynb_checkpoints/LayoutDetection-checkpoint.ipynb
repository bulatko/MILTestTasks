```python
import json
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import ImageDraw, Image
import segmentation_models_pytorch as smp
import torch
from torch import nn
import albumentations as A
from torch.utils.data import DataLoader, Dataset
from torchvision.transforms import ToTensor
```

```python
train = json.load(open('train.json', 'r'))
test = json.load(open('test.json', 'r'))
```

```python
# Функция для перевода многоугольника в маску
# Похоже, для трейна и теста сосо файлы отличаются в тесте не нормированы координаты вершин :/
def polygon_to_mask(xy, w, h, c, normed_dists=True):
  xy = np.array(xy)
  xy = np.array(xy[0])
  if normed_dists:
    xy[::2] *= w
    xy[1::2] *= h
  xy = xy.reshape(-1, 2)
  xy = [tuple(a) for a in xy]
  img = Image.new('L', (w, h), 0)
  if len(xy) < 2:
    mask = np.array(img)
    return mask

  ImageDraw.Draw(img).polygon(xy, outline=c, fill=c)
  mask = np.array(img)
  return mask

# Функция, которая по всем аннотациям получает полноценную маску с классами
def get_mask(image, annots, normed_dists=True):
  w, h = image['width'], image['height']
  mask = np.zeros((h, w))
  for a in annots:
    data = polygon_to_mask(a['segmentation'], w, h, a['category_id'], normed_dists)
    mask = np.where(data, a['category_id'], mask)
  return mask
```

```python
# Посмотрим, как у нас всё работает
ind = np.random.randint(0, len(test['images']))
image = test['images'][ind]
im_id = image['id']
annots = [a for a in test['annotations'] if a['image_id'] == im_id]
_, ax = plt.subplots(1, 2, figsize=(20, 10))
ax[0].imshow(plt.imread(os.path.join('data', image['file_name'])))
ax[1].imshow(get_mask(image, annots, False))
```

```python
# Напишем свой датасет (учтем ненормированный тест)
class PapersDataset(Dataset):
  def __init__(self, images, annotations, transforms=None, normed_dists=True):
    self.images = images
    self.annotations = []
    self.normed_dists = normed_dists
    for im in images:
      im_id = im['id']
      annots = [a for a in annotations if a['image_id'] == im_id]
      self.annotations.append(annots)
    self.transforms = transforms

  def __len__(self):
    return len(self.images)

  def __getitem__(self, i):

    image = self.images[i]
    annots = self.annotations[i]
    mask = get_mask(image, annots, self.normed_dists)
    image = plt.imread(os.path.join('data', image['file_name']))
    if self.transforms:
      res = self.transforms(image=image, mask=mask)

    image, mask = res['image'], res['mask']
    image = ToTensor()(image)
    mask = ToTensor()(mask)
    return image, mask[0].long()
```

```python
# Добавим аугментаций, которые на выходе будут давать что-то всё еще похоже на пейпер
# Подожесть смотрел тут: https://albumentations-demo.herokuapp.com/
train_transforms = A.Compose([
                              A.Resize(256, 256),
 #                             A.OneOf([
 #                                      A.RandomCrop(256, 256),
 #                                     A.Resize(256, 256)
 #                             ], 1),
 #                             A.CLAHE(),
 #                             A.Blur(),
 #                             A.ChannelShuffle(),
 #                             A.RandomGridShuffle(),
 #                             A.RandomBrightness(),
 #                             A.ShiftScaleRotate(0.5, 0.1, 0),
                              A.Normalize()
])
# В итоге лучше показал результат без аугментаций

# Для теста попроще, ресайз + нормализация для претрейн модельки
test_transforms = A.Compose([
                             A.Resize(256, 256),
                             A.Normalize()
])

train_dataset = PapersDataset(train['images'], train['annotations'], train_transforms)
test_dataset = PapersDataset(test['images'], test['annotations'], test_transforms, False)
```

```python
# Посмотрим на результаты датасета
ind = np.random.randint(0, len(test_dataset))
im, mask = test_dataset[ind]
im = im.permute(1, 2, 0)
_, ax = plt.subplots(1, 2)
ax[0].imshow(im)
ax[1].imshow(mask)
```

```python
# Даталоадеры с двумя воркерами (на колабе осталось 2)
train_dataloader = DataLoader(train_dataset, 8, True, num_workers=2)
test_dataloader = DataLoader(test_dataset, 8, num_workers=2)
```

```python
# Возьмём SOTA semantic segmentation модельку на одном из SOTA бэкбонов из библиотеки segmentation_models.pytorch
model = smp.DeepLabV3Plus('efficientnet-b7', encoder_weights='imagenet', classes=3)
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
```

```python
# Самый оптимальный выбор лосса после эксперементов - выдал результат лучше жаккарова лосса и кросс энтропии
loss = smp.losses.DiceLoss('multiclass')
loss.__name__ = "Loss"

# Подправим IOU из коробки smp
class IoU(smp.utils.base.Metric):
    __name__ = 'iou_score'

    def __init__(self, eps=1e-7, threshold=0.5, activation=None, ignore_channels=None, **kwargs):
        super().__init__(**kwargs)
        self.eps = eps
        self.threshold = threshold
        self.ignore_channels = ignore_channels

    def forward(self, y_pr, y_gt):

        y_pr = nn.functional.softmax(y_pr)

        y_gt = nn.functional.one_hot(y_gt, 3).permute(0, 3, 1, 2)
        return smp.utils.functional.iou(
            y_pr, y_gt,
            eps=self.eps,
            threshold=self.threshold,
            ignore_channels=self.ignore_channels,
        )

metrics = [
  IoU(threshold=0.5),
]

optimizer = torch.optim.Adam([ 
    dict(params=model.parameters(), lr=0.0006),
])
```

```python
# Trainloop'ы по эпохам тоже возьмем из их коробки
train_epoch = smp.utils.train.TrainEpoch(
    model, 
    loss=loss, 
    metrics=metrics, 
    optimizer=optimizer,
    device=device,
    verbose=True,
)

valid_epoch = smp.utils.train.ValidEpoch(
    model, 
    loss=loss, 
    metrics=metrics, 
    device=device,
    verbose=True,
)
```

```python
# Учим
test_max_score = 0
train_max_score = 0

for i in range(0, 40):
    print('\nEpoch: {}'.format(i))
    train_logs = train_epoch.run(train_dataloader)
    valid_logs = valid_epoch.run(test_dataloader)
    
    # Чекпоинт модели с лучшим mIoU на тесте
    if test_max_score < valid_logs['iou_score']:
        test_max_score = valid_logs['iou_score']
        torch.save(model, 'model.pth')
        print('Model saved!')

    if train_max_score < train_logs['iou_score']:
        train_max_score = train_logs['iou_score']
    
    optimizer.param_groups[0]['lr'] *= 0.93
    
```

```python
print("Best train mIOU = {:.4%}".format(train_max_score))
print("Best test mIOU = {:.4%}".format(test_max_score))
```
